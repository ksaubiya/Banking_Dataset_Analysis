# Visualization:-

### Pair Plot

```python
#Pair plot use for when you want to relationship between two variables
sns.pairplot(payment)
```

### HeatMap:-

```python
corelation=payment.corr()
```

```python
sns.heatmap(corelation, xticklabels=corelation.columns,yticklabels=corelation.columns,
            annot=True)
```

### Scatter Plot:- 

```python
sns.relplot(x='year',y='month_number',hue='month',data=payment)
```

### Histrogram:- 

```python
sns.distplot(payment['month_number'])
```

### Box Plot:- 

```python
sns.catplot(x='month_number',kind='box',data=payment)
```

```python
sns.catplot(x='year',kind='box',data=payment)
```

### Joint Plot:-

```python
sns.jointplot(x='month',y='month_number',data=payment,height=10,ratio=3,color='pink')
```
